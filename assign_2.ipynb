```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/House_Price_India.csv")
dataset.head()
```

# Univariate analysis


```python
uni_var = 'Price'
print(dataset[uni_var].describe())
```

#### Histogram


```python
plt.figure(figsize=(8, 5))
plt.hist(dataset[uni_var], bins=20, color='skyblue', edgecolor='black')
plt.title(f'Distribution of {uni_var}')
plt.xlabel(uni_var)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=dataset, x=uni_var, bins=30, kde=True, color='skyblue')
plt.title(f'Distribution of {uni_var}')
plt.xlabel(uni_var)
plt.ylabel('Frequency')
plt.show()
```

#### Box plot


```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=dataset, x=uni_var, color='lightblue')
plt.title(f'Box Plot of {uni_var}')
plt.xlabel(uni_var)
plt.show()
```

# Bivariate Analysis

```python
var_x = "number of bedrooms"
var_y = "living area"
```

#### Line plot

```python
plt.figure(figsize=(8, 5))
sns.lineplot(data=dataset, x=var_x, y=var_y, marker='o', color='blue')
plt.title('Bivariate Line Plot with Seaborn')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.grid(True)
plt.show()
```

#### Scatterplot


```python
#matplotlib
plt.figure(figsize=(8, 5))
plt.scatter(dataset[var_x], dataset[var_y], color='skyblue', alpha=1)
plt.title(f'{var_x} vs. {var_y}')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.show()
```

```python
#seaborn
plt.figure(figsize=(8, 5))
sns.scatterplot(data=dataset, x=var_x, y=var_y, color='skyblue', alpha=1)
plt.title(f'{var_x} vs. {var_y}')
plt.xlabel(var_x)
plt.ylabel(var_y)
plt.show()
```

# Mulitvariate Analysis

#### Heatmap

```python
var_z = "Price"

var_corr = [var_x, var_y, var_z]
correlation_matrix = dataset[var_corr].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1)
plt.title(f'Correlation Heatmap between {", ".join(var_corr)}')
plt.show()
```

#### Pair plot

```python
sns.pairplot(data=dataset, vars=[var_x, var_y, var_z], diag_kind='kde', kind='scatter')
plt.suptitle(f'Pair Plot of {var_x}, {var_y}, and {var_z}')
plt.show()
```
